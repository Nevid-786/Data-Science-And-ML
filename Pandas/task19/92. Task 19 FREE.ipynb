### `Q-4:` Perform `groupby` based on the `"Team"` column and then perform Z Normalization on top of the below columns of each group:
1. Passes
2. Passes Completed
3. Attempted Line Breaks
4. Completed Line Breaks

You have to make a python function named `z_normalization` which takes two arguments:

1. *group:* Every group that you have created
2. *cols_to_perform:* This parameter takes a list of columns on which you have to perform the Z-Normalization.

For this problem, you have to use th `apply()` method.

$\Large Z \ - \ Normalization = \frac{X_i - \mu}{std} $

After that find out the below values for each group:
- minimum "Passess"
- maximum "Passess"
- minimum "Yellow Cards"
- maximum "Yellow cards"
- average "Yellow Cards"
- maximum "Attempted Line Breaks"
- minimum "Attempted Line Breaks"
- standardard deviation of "Attempted Line Breaks"
- average Possession

````python
import pandas as pd

fifa_df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vT3D_x_4DS6d51LKJ7ze1sxT5WpV5uiSVOFYHLwBiGru6vFyVv5h5-83AwFjxWYiWfCDjDAaarHAV-k/pub?gid=0&single=true&output=csv")
````

````python
fifa_df.head(3)
````

````python
grouped_fifa=fifa_df.groupby("Team",group_keys=False)
grouped_fifa.groups.keys()
````

````python
def z_normalization(group,columns=[]):
    for column in columns:
        group[column]=(group[column]-group[column].min())/(group[column].max()-group[column].min())
    return group
````

````python
grouped_fifa.apply(z_normalization,columns=["Passes","Passes Completed"],include_groups=True)
````

````python
# - minimum "Passess"
# - maximum "Passess"
# - minimum "Yellow Cards"
# - maximum "Yellow cards"
# - average "Yellow Cards"
# - maximum "Attempted Line Breaks"
# - minimum "Attempted Line Breaks"
# - standardard deviation of "Attempted Line Breaks"
# - average Possession
grouped_fifa.agg({
    "Passes":["min","max"],
    "Yellow Cards":["min","max","mean"],
    "Attempted Line Breaks":["min","max","std"],
    "Possession (%)":"mean"
}).head()
````

###`Q-5:` Find batsman in below category-
* Highest score while chasing
* Best Strike rate while chasing and have faced 100+ balls


> Chasing mean team batting in second inning

````python
# code here

#batting in 2nd inning=chasing acc to individual matches

Chasing_Matches=ipl[(ipl["inning"]==2)]
ipl_group=Chasing_Matches.groupby(["match_id","batsman"])
ipl_group["batsman_runs"].sum().sort_values(ascending=False).reset_index().iloc[0]["batsman"]

    


````

````python
#chasing and have faced 100+ balls in 2nd inning

Chasing_no_wide_balls=ipl[(ipl["inning"]==2) & (ipl["wide_runs"]==0)]

````

````python
batsman_balls=Chasing_no_wide_balls["batsman"].value_counts()


ipl_group=Chasing_no_wide_balls.groupby(["match_id","batsman"])
ipl_group.apply(lambda x:(x["batsman_runs"].sum()/x["batsman_runs"].count() if batsman_balls[x.name[1]]>100 else 0)*100).sort_values(ascending=False).head(2)
````
