No. of rows using apply function on grpby object

```python
def Count(data):
    # return data.shape[0]
    return data
grouped.apply(Count)
```

No. of movies starts with A in Each Gener

```python
def function(data):
    return data["Series_Title"].str.startswith("A").sum()
grouped.apply(function)
```

Rank of Movies in Each Gruop of Gener acc to withn grps

```python
def function(data):
    data["Grp Ranks"]=data["Rating"].rank(ascending=False)
    
    return data
grouped.apply(function,include_groups=False).reset_index()
```
