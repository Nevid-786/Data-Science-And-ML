```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset("tips")
df
```

```python
smoker_male = df[(df["sex"] == "Male") & (df["smoker"] == "Yes")]
smoker_female = df[(df["sex"] == "Female") & (df["smoker"] == "Yes")]
non_smoker_male = df[(df["sex"] == "Male") & (df["smoker"] == "No")]
non_smoker_female = df[(df["sex"] == "Female") & (df["smoker"] == "No")]
```

```python
plt.figure(figsize=(8,6))

plt.scatter(smoker_male["total_bill"], smoker_male["tip"],
            color="red", label="Male Smoker")

plt.scatter(smoker_female["total_bill"], smoker_female["tip"],
            color="blue", label="Female Smoker")

plt.scatter(non_smoker_male["total_bill"], non_smoker_male["tip"],
            color="green", label="Male Non-Smoker")

plt.scatter(non_smoker_female["total_bill"], non_smoker_female["tip"],
            color="orange", label="Female Non-Smoker")

plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Total Bill vs Tip")
plt.legend()
plt.grid(True)
plt.show()
```
