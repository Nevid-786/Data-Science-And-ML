```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
with open("smart_phone.html","r",encoding="utf-8") as f:
    content = f.read()
    soup=BeautifulSoup(content,"lxml")
```

```python
containers=soup.find_all("div",{"class":"sm-product has-tag has-features has-actions"})
```

```python
names = []
prices = [] 
ratings = []
sim = []
processor = []
ram = []
battery = []
display = []
camera = []
card = []
os = []


```

```python

def get_data(tag, *attrs):
   try:
        if tag is None:
            raise ValueError("None")

        for attr in attrs:
            
            tag = getattr(tag, attr)
            if tag is None:
                raise ValueError("None value")
      
        return tag
   except Exception as e:
       return np.nan

    


```

```python
def get_specs(spec_list,index):
    try:
        print(len(spec_list))
        item=spec_list[index]
        if item is None:
            raise ValueError("None")
        
        return item
    except Exception as e:
        print(e)
        return np.nan
        
    
  
```

```python
for x in containers:
   specs = x.find("ul", class_="sm-feat specs").find_all("li")

   sim_spec = processor_spec = ram_spec = battery_spec = np.nan
   display_spec = camera_spec = card_spec = os_spec = np.nan

   for li in specs:
      text = li.get_text(strip=True)

      if "Sim" in text or "VoLTE" in text:
         sim_spec = text

      elif "Processor" in text or "Snapdragon" in text or "Dimensity" in text:
         processor_spec = text

      elif "RAM" in text:
         ram_spec = text

      elif "mAh" in text:
         battery_spec = text

      elif "Display" in text or "Hz" in text:
         display_spec = text

      elif "Camera" in text:
         camera_spec = text

      elif "Memory Card" in text:
         card_spec = text

      elif "Android" in text or "iOS" in text:
         os_spec = text

   sim.append(sim_spec)
   processor.append(processor_spec)
   ram.append(ram_spec)
   battery.append(battery_spec)
   display.append(display_spec)
   camera.append(camera_spec)
   card.append(card_spec)
   os.append(os_spec)
   names.append( get_data(
   x.find("a",class_="name clamp-2"),
   "h2",
   "text"))

   prices.append(get_data(x.find("span",class_="price"),"text"))
   ratings.append(get_data(x.find("div",class_="score"),"b","text"))
   
```

```python
df =pd.DataFrame({
    "Name":names,
    "Price":prices,
    "Rating":ratings,
    "Sim":sim,
    "Processor":processor,
    "RAM":ram,
    "Battery":battery,
    "Display":display,
    "Camera":camera,
    "Memory Card":card,
    "OS":os
})
df.to_csv("smart_phone.csv",index=False,encoding="utf-8-sig")
```
