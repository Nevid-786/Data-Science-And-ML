```python
df= main_df.copy()
```

```python
df.info()
```

```python
#Price
df['Price']=df['Price'].str.replace('₹','').str.replace(',','').astype(float)
```

```python
df.fillna("None",inplace=True)
df.shape
```

```python
df.drop(df[~df["Sim"].str.contains("G")].index, inplace=True)
df.shape
```

```python
temp_df=df[["Name"]]
df["Name"]=temp_df["Name"].str.split("(").apply(lambda x: x[0])
df["Name"]=df["Name"].str.replace("4G","").str.replace("5G","").str.replace("  "," ")
```
